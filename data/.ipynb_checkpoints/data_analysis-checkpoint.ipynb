```python
import pandas as pd
import seaborn as sns
import numpy as np
sns.set(rc = {'figure.figsize':(15,10)})

```

```python
# load data
df = pd.read_pickle('D:/Masters/NS_EoS/dataset/NS_EoS.pkl')
```

```python
#basic info
print(df.info())
print(df.describe())
df.head()
```

```python
#data structure
n_measurements = len(df.index)
n_unique = len(df.drop_duplicates(df.columns[:7]))
n_stars = len(df.groupby(['Mass','Radius']))
n_eos = len(df.groupby(['m1','m2']))

print('Total number of measurements:',n_measurements)
print('Number of duplicate measurements:',(n_measurements - n_unique))
print('Number of stars:',n_stars)
print('Number of EoS parameters:',n_eos)
print('Average number of stars per EoS parameter:',(n_stars/n_eos))
print('Average number of NP readings per star:',n_unique/n_stars)
```
